```python
import numpy
import scipy.stats as ss
import matplotlib.pyplot as plt
import seaborn as sns
import math

%matplotlib inline
```

```python
def get_plot(npoints, points=1000):
    x = numpy.random.rand(npoints)
    xg = ss.gaussian_kde(x)
    xk = numpy.linspace(numpy.min(x), numpy.max(x), points)
    xgv = xg(xk)
    return xk, xg(xk)
```

#### < 1000 points

```python
points_lst = list(range(1100)[100::100])
plots = [get_plot(npoints) for npoints in points_lst]
```

```python
N_COLUMNS = 5

fig, axs = plt.subplots(
    math.ceil(len(plots)/N_COLUMNS),
    N_COLUMNS,
    figsize=(12, 6)
)

axs = axs.flatten()

for i, plot in enumerate(plots):
    ax = axs[i]
    ax.plot(*plot)
    ax.set_title('%d points' % points_lst[i])

plt.tight_layout()
```

```python
plt.figure(figsize=(15, 10))
for i, plot in enumerate(plots):
    plt.plot(*plot, label='%d points' % points_lst[i])
plt.legend()
plt.title('all together')
```

```python
nplots = len(plots)

mat = numpy.zeros((nplots, nplots))

for r in range(nplots):
    for c in range(nplots):
        mat[r][c] = ss.entropy(plots[r][1], plots[c][1])

plt.figure(figsize=(12, 8))
g = sns.heatmap(mat, annot=True)
labels = ['%d points' % n for n in points_lst]
g.set_xticklabels(labels[::-1])
g.set_yticklabels(labels)
for xtick, ytick in zip(g.get_xticklabels(), g.get_yticklabels()):
    xtick.set_rotation(45); ytick.set_rotation(45);
g.set_title('KL divergence matrix')
```
